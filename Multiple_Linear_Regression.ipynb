```python
#Import pandas library

import pandas as pd
```

```python
#Read the dataset from csv file

data_all = pd.read_csv('Boston.csv')
```

```python
# Have a glance at the shape
# Print first five rows of the data

data_all
```

```python
# Set up the dependent (labels) and the independent (features) variables
 
X_features = pd.DataFrame(data_all.iloc[:,:-1])
y_labels = pd.DataFrame(data_all.iloc[:,-1])
```

```python
# Have a glance at the independent variable

X_features
```

```python
# Have a glance at the dependent variable

y_labels
```

```python
# Split the data into train and test sets

from sklearn.model_selection import train_test_split

# 60% for training set, 40% for test set 
X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels, test_size= 0.4, random_state=1)


# Data for test set will be divided into two. 20% for test set and 20% for validation set
# X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=1)


# To print out the all data percentage. 
data = [y_train, y_test]
data_labels = ["Train: ", "Test: "]


for (data_label, dataset) in zip(data_labels,data):

    print ( (data_label) , round(len(dataset)/len(y_labels), 2))
```

```python
 # Have a glance at the shape of the train and test sets:
 
 # Shape of the train,  test and validation sets

print()
print ("X_train:", X_train.shape)
print ("y_train:", y_train.shape)
print ("X_test:", X_test.shape)
print ("y_test:", y_test.shape)
# print ("X_val:", X_val.shape)
# print ("y_val:", y_val.shape)
```

```python
# Train the algorithm

from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)  # # create object for the class

linear_regressor.fit(X_train, y_train) # perform linear regression

# X_train.head(10)
```

```python
# Having a look at the coefficients that the model has chosen:

v = pd.DataFrame(linear_regressor.coef_ , index = ["Co-efficient"]).transpose()
w = pd.DataFrame(X_features.columns, columns=["Attribute"])
```

```python
# Concatenating the DataFrames to compare:

coeff_df = pd.concat([w, v], axis=1, join='inner')

coeff_df
```

```python
# Comparing the predicted value to the actual value:

# Predicted value

y_pred = linear_regressor.predict(X_test)
y_pred = pd.DataFrame(y_pred, columns=["Predicted"])

y_pred
```

```python
y_test
```

```python
# Evaluate algorithm

from sklearn import metrics
import numpy as np

print ('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_pred))
print ('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_pred))
print ('Root Mean Squared Error: ', np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
```
